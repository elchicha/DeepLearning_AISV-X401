```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Generate input data
x = torch.linspace(-10, 10, 100)

# Activation functions
relu = torch.nn.functional.relu(x)
sigmoid = torch.sigmoid(x)
tanh = torch.tanh(x)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), relu.numpy(), label='ReLU')
plt.plot(x.numpy(), sigmoid.numpy(), label='Sigmoid')
plt.plot(x.numpy(), tanh.numpy(), label='Tanh')
plt.title("Activation Functions in PyTorch")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

## Notes on Activation Functions in Machine Learning

Activation functions are a key component in neural networks. They introduce non-linearity into the model, allowing it to learn complex patterns in data. Here are the three common activation functions plotted above:

- **ReLU (Rectified Linear Unit):** Outputs zero for negative inputs and the input itself for positive values. It is widely used due to its simplicity and effectiveness in deep networks.
- **Sigmoid:** Maps input values to a range between 0 and 1. It is useful for probabilistic interpretations but can suffer from vanishing gradients for very large or small inputs.
- **Tanh:** Maps input values to a range between -1 and 1. Like sigmoid, it can also suffer from vanishing gradients but is zero-centered, which can help optimization.

The plot above shows how each activation function transforms the input values. Choosing the right activation function can significantly impact the performance and convergence of a neural network.

```python
# Compute softmax activation for a sample vector
logits = torch.tensor([2.0, 1.0, 0.1])
softmax_vals = torch.nn.functional.softmax(logits, dim=0)

# Plotting other activations
plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), relu.numpy(), label='ReLU')
plt.plot(x.numpy(), sigmoid.numpy(), label='Sigmoid')
plt.plot(x.numpy(), tanh.numpy(), label='Tanh')
plt.title("Activation Functions in PyTorch")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()

# Plotting softmax for sample logits
plt.figure(figsize=(6, 4))
plt.bar(['Class 1', 'Class 2', 'Class 3'], softmax_vals.numpy())
plt.title("Softmax Output for Example Logits")
plt.ylabel("Probability")
plt.show()
```

### Softmax Activation Function

The **softmax** activation function is commonly used in the output layer of neural networks for multi-class classification problems. It converts a vector of raw scores (logits) into probabilities, where each value is in the range (0, 1) and the sum of all probabilities is 1.

Mathematically, for an input vector \( x = [x_1, x_2, ..., x_n] \), the softmax function is defined as:

\[
\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}}
\]

This means each output value depends on all input values, making softmax sensitive to the relative scale of the inputs. It is especially useful for interpreting the output of a neural network as a probability distribution over classes.
